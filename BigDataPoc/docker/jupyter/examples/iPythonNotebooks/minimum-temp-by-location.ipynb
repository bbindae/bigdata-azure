```python
from pyspark import SparkContext, SparkConf
conf = SparkConf().setMaster("local").setAppName("MinimumTempByLocation")
sc = SparkContext(conf = conf)
```

```python
lines = sc.textFile("/home/jovyan/examples/data-samples/spark-examples/1800.csv")
```

```python
def parseLine(line):
    fields = line.split(',')
    stationId = fields[0]
    entryType = fields[2]
    temperature = fields[3] #float(fields[3] * 0.1 * (9.0 / 5.0) + 32.0)
    return (stationId, entryType, temperature)

parsedLines = lines.map(parseLine)
```

```python
for x in parsedLines.collect():
    print("stationId: %s    temp: %s" % (x[0], str(x[2])))
```

```python
minTemps = parsedLines.filter(lambda x: "TMIN" in x[1])
stationTemps = minTemps.map(lambda x: (x[0], x[2]))
```

```python
minTemps = stationTemps.reduceByKey(lambda x, y: min(x, y))
results = minTemps.collect()

for result in results:
    print("%s: %s" % (result[0], result[1]))
```
