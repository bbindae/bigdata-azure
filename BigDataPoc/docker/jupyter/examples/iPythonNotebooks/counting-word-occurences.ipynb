```python
from pyspark import SparkContext, SparkConf

conf = SparkConf().setMaster("local").setAppName("CountingWordOccurences")
sc = SparkContext(conf = conf)
```

```python
input = sc.textFile("/home/jovyan/examples/data-samples/spark-examples/Book.txt")
```

```python
import re
def normalizeWords(text):
    return re.compile(r'\W+', re.UNICODE).split(text.lower())
```

```python
words = input.flatMap(normalizeWords)
for w in words.collect():
    print(w)
```

```python
results = words.countByValue()
for word, count in results.items():
    cleanWord = word.encode('ascii', 'ignore')
    if(cleanWord):
        print("%s %i" % (cleanWord, count))
```

```python
wordsCount = input.flatMap(normalizeWords).map(lambda x: (x,1)).reduceByKey(lambda x, y: x + y)
```

```python
for x in wordsCount.collect():
    print("%s %i" % (x[0], x[1]))
```

```python
wordCountsSorted = wordsCount.map(lambda x: (x[1],x[0])).sortByKey()
results = wordCountsSorted.collect()
for result in results:
    count = str(result[0])
    word = result[1].encode('ascii', 'ignore')
    if(word):
        print("%s: \t\t %s" % (word, count))
```
